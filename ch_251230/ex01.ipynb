```python
import numpy as np
import pandas as pd
import os

import chardet

# 서울 도서관 11월 대출 현황 데이터
file_name = "seoul_library.csv"
path = f'{os.getcwd()}/{file_name}'

if os.path.exists(path):
    
    with open(path, mode = "rb") as f:
        encoding_type = chardet.detect(f.readline())
        
    df = pd.read_csv(path, encoding = encoding_type["encoding"])
    # load_counts = df[['도서명']].value_counts()
    # top10_books = load_counts.head(10)
    # top10_df = top10_books.reset_index()
    # top10_df.columns = ['도서명', '대출건수']
    # display(top10_df)
    
    # path = f'{os.getcwd()}/{"top10_books.csv"}'
    # top10_df.to_csv(path, encoding = encoding_type["encoding"])
    # loan_counts = df.groupby("주제분류번호").agg(np.sum, np.mean)
    # ].value_counts() #| 분류 | 대출건수합계 | 평균대출건수 |
    # df.groupby('분류컬럼명')
    # df = load_counts.agg(['sum', 'mean'])
    # result = df.groupby('주제분류번호')['loan_count'].agg(['sum', 'mean'])
    
    # 데이터 결측값 / 타입 정리
    df['대출건수'] = pd.to_numeric(df['대출건수'], errors = 'coerce').fillna(0).astype(int)
    result = df.groupby('주제분류번호').agg(대출건수합계=('대출건수', 'sum'), 평균대출건수=('대출건수', 'mean'))

    print(result)
```
